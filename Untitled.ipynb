```python
import pandas as pd
```

```python
dfReturns = pd.concat([samsungElec['log_return'], KGMobil['log_return'], hyosungChem['log_return'], celltrion['log_return']], axis=1, join='inner')
dfReturns.columns = ['삼성전자', 'KG 모빌리언스', '효성화학', '셀트리온']
```

```python
dfReturns = dfReturns[1:]
dfReturns.head(10)
```

```python
volatilities = dfReturns.std() * np.sqrt(252)
```

```python
volatilities
```

```python
pd.DataFrame(volatilities.T)
```

```python
capSE = 5561791
capKG = 4290
capHS = 12761
capCell = 396586
capSum = capSE + capKG + capHS + capCell
cap_weights = [capSE/capSum, capKG/capSum, capHS/capSum, capCell/capSum]
dfCW = pd.DataFrame(cap_weights).T
dfCW.columns = ['삼성전자', 'KG 모빌리언스', '효성화학', '셀트리온']
dfCW.index = ['cap_weights']
dfCW
```

```python
# Q
views = [0.079, 0.049]
# P
pick_list = [
        {
            "효성화학": 1
        },
        {
            "KG 모빌리언스": 1
        }
]
```

```python
stocks = ['삼성전자', 'KG 모빌리언스', '효성화학', '셀트리온']
```

```python
correlation = pd.DataFrame(correlation, index=stocks, columns=stocks)
```

```python
volatilities = pd.DataFrame(volatilities,
                            index=stocks, columns=["vol"])
covariance = volatilities.dot(volatilities.T) * correlation
```

```python
cap_weights = pd.DataFrame(cap_weights,
                              index=stocks, columns=["CapWeight"])
```

```python
from model import *
```

```python
bl = BlackLitterman()
bl.allocate(covariance=covariance,
            market_capitalised_weights=cap_weights,
            investor_views=views,
            pick_list=pick_list,
            asset_names=covariance.columns,
            tau=0.05,
            risk_aversion=2.5)
```

```python
print(bl.weights)
```

```python
pd.DataFrame(bl.weights.values, columns=['삼성전자', 'KG 모빌리언스', '효성화학', '셀트리온'], index=['weights'])
```

```python
bl.implied_equilibrium_returns['귀속 초과 균형 수익률'] = '귀속 초과 균형 수익률'
bl.implied_equilibrium_returns.set_index('귀속 초과 균형 수익률', inplace=True)
bl.implied_equilibrium_returns.index.name=None
bl.implied_equilibrium_returns
```

```python
bl.posterior_expected_returns['사후 수익률'] = '사후 수익률'
bl.posterior_expected_returns.set_index('사후 수익률', inplace=True)
bl.posterior_expected_returns.index.name=None
bl.posterior_expected_returns
```

```python
cap_weights.T
```

```python
er = bl.implied_equilibrium_returns.T
pr = bl.posterior_expected_returns.T

compare = pd.concat([er, pr], axis=1)
```
